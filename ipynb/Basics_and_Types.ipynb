## 1. Базовые понятия
`Python` — интерпретируемый язык. Дефолтный интерпретатор — **CPython**.

**Как он работает:**
1. Преобразует исходный код в `bytecode`.
2. Исполняет его с помощью стека и окружения (словарь с переменными).

```python
import dis

def example_calculation():
    x = 30
    y = 62
    z = x + y
    return z

# Посмотрим на байткод в реальном времени
dis.dis(example_calculation)
```

### Динамическая типизация
Python — динамический язык. Проверка типов происходит во время вызова.

```python
def add(a, b):
    return a + b

def f():
    return "This string will be replaced by logic"

# 'Everything is an object': подменяем байткод функции
f.__code__ = add.__code__ 
print(f"Результат f(62, 30): {f(62, 30)}")
```

## 2. Встроенные типы

### None, Bool и Числа

```python
# None
res = print("None check")
print(f"print() returns: {res}")
print(f"res is None: {res is None}")

# Bool & Short-circuit
print("\n--- Logic ---")
False and print("This won't print")
True and print("This will print")

# Numbers
print("\n--- Numbers ---")
print(f"True + True: {True + True}")
print(f"Big int: {2 ** 128}")
print(f"Float division: {5 / 2}")
print(f"Floor division: {5 // 2}")
print(f"Negative floor: {-3 // 2}") # Округляет вниз
```

## 3. Коллекции

### List, Set, Dict

```python
# Списки и срезы
xs = list(range(5))
print(f"Original list: {xs}")
xs[1:-1:2] = [999, 888] 
print(f"After slice assignment: {xs}")

# Словари (Dict)
d = {'year': 2026, 'month': 'April'}
print(f"Keys: {d.keys()}")
print(f"Safe get: {d.get('day', 11)}")
```

## 4. Изменяемость (Mutability)
Важнейшая концепция: переменная — это просто ссылка на объект в памяти.

```python
print("--- Mutable (List) ---")
a = [1, 2, 3]
b = a
b.append(4)
print(f"a: {a} (изменился вместе с b)")

print("\n--- Immutable (Int) ---")
x = 10
y = x
x += 5
print(f"x: {x}, y: {y} (y остался прежним)")
```
