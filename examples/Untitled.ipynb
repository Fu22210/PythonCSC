```python
import time
```

```python
def time_deco(func: Callable):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"время выполнения функции {end - start}")
        return res
    return wrapper 
```

```python
@time_deco
def g(sec=1):
    time.sleep(sec)
    return 123
```

```python
def limit_calls(limit: int):
    def wrapper(func: Callable):
        @wraps(func)
        def inner(*args, **kwargs):
            start = time.time()
            nonlocal limit
            if limit == 2:
                print( "ERROR")
            res = func(*args, **kwargs)
            end = time.time()
            print(f"время выполнения функции {end - start}")
            return res
        return inner
    return wrapper
```

```python
@limit_calls(1)
def func(sec: int =1):
    time.sleep(sec)
    return 123
```
